```python
import pandas as pd 
```

```python
data=pd.read_csv(r"C:\Users\wwrao\Downloads\cleaned_file.csv")
data.head()
```

```python
dataset=data[["Gender","Married","Education","Self_Employed","Property_Area","Loan_Status"]]
dataset.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dataset.drop(columns=["Loan_Status"],axis=1),dataset["Loan_Status"],test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
x_train
```

```python
x_test.shape
```

```python
y_train.shape,y_test.shape
```

```python


transform = ColumnTransformer(
    transformers=[
        ("sami1", OneHotEncoder(drop="first"), ["Gender"]),
        ("sami2", OneHotEncoder(drop="first"), ["Married"]),
        ("sami3", OneHotEncoder(drop="first"), ["Education"]),
        ("sami4", OneHotEncoder(drop="first"), ["Self_Employed"]),
        ("sami5",OrdinalEncoder(categories=[["Rural",'Semiurban',"Urban"]]),["Property_Area"])
        
        
    ],
    remainder="drop"
)

```

```python
transform
```

```python
x_train_scaled=transform.fit_transform(x_train)
x_test_scaled=transform.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
x_train["Gender"].value_counts()
```

```python
df1=pd.DataFrame(x_train_scaled,columns=["Gender","Married","Education","Self_Employed","Property_Area"])
```

```python
df2=pd.DataFrame(x_test_scaled,columns=["Gender","Married","Education","Self_Employed","Property_Area"])
```

```python
df=pd.concat((df1,df2),axis=0,ignore_index=True)
df
```
