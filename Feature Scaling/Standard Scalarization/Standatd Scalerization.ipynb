```python
import pandas as pd 
```

```python
dataset=pd.read_csv(r"C:\Users\wwrao\Downloads\income.csv")
dataset.head()
```

```python
dataset["Income($)"].isnull().sum()
```

```python
import matplotlib.pyplot as plt 
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(5,2))

ax1.plot(dataset["Age"])
ax1.set_title("Age")
ax1.set_xlabel("Age")

ax2.plot(dataset["Income($)"])
ax2.set_title("Salary")
ax2.set_xlabel("Salary")


plt.tight_layout()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(dataset[["Income($)"]])
```

```python
dataset["Income"]=ss.fit_transform(dataset[["Income($)"]])
dataset.head()
```

```python
dataset["Income"].describe()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(5,2))



plt.tight_layout()
ax1.plot(dataset["Income($)"])
ax1.set_title(" before Scale Salary")
ax1.set_xlabel("Salary")


ax2.plot(dataset["Income"])
ax2.set_title("after Scale")
ax2.set_xlabel("Salary")
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(5,3))

ax1.scatter(dataset["Age"],dataset["Income($)"])
ax1.set_title(" before Scale Salary")
ax1.set_ylabel("Salary")
ax1.set_xlabel("Age")

ax2.scatter(dataset["Age"], dataset["Income"])
ax2.set_title("after Scale")
ax2.set_ylabel("Salary")
ax2.set_xlabel("Age")
plt.tight_layout()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(5,2))



plt.tight_layout()
ax1.hist(dataset["Income($)"])
ax1.set_title(" before Scale Salary")
ax1.set_xlabel("Salary")


ax2.hist(dataset["Income"])
ax2.set_title("after Scale")
ax2.set_xlabel("Salary")
```
